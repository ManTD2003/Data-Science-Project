```python
import requests

def get_sellerid(product_id,sp_id):
    url = f'https://tiki.vn/api/v2/products/{product_id}'
    try:

        headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
                'Accept': 'application/json',
                # Thêm các tiêu đề khác nếu cần, ví dụ như Authorization
            }
        params = {
                'platform': 'web',
                'spid': sp_id,
                'version':3
            }
        # Gửi yêu cầu GET tới API
        response = requests.get(url,params=params,headers=headers)
        # Kiểm tra mã trạng thái của phản hồi
        response.raise_for_status()
        
        # Xử lý dữ liệu phản hồi
        data = response.json() 
        if data['current_seller'] == None or data['current_seller']['id'] == None:
            return 1
        else:
            return data['current_seller']['id']
      # In dữ liệu ra hoặc xử lý tùy ý
    except requests.exceptions.RequestException as e:
        # Xử lý lỗi
        print("Có lỗi xảy ra:", e)
```

```python
import requests

def get_comment(product_id,sp_id,seller_id,page=1,limit=5):

    url = 'https://tiki.vn/api/v2/reviews'
    try:

        headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
                'Accept': 'application/json',
                # Thêm các tiêu đề khác nếu cần, ví dụ như Authorization
            }
        params = {
                'limit': limit,
                'page': page,
                'include': 'comments,contribute_info,attribute_vote_summary',
                'product_id': product_id,
                'sp_id': sp_id,
                'seller_id': seller_id
            }
        # Gửi yêu cầu GET tới API
        response = requests.get(url,params=params,headers=headers)
        # Kiểm tra mã trạng thái của phản hồi
        response.raise_for_status()
        
        # Xử lý dữ liệu phản hồi
        data = response.json() 
        return data["data"]
      # In dữ liệu ra hoặc xử lý tùy ý
    except requests.exceptions.RequestException as e:
        # Xử lý lỗi
        print("Có lỗi xảy ra:", e)
```

```python
import requests

def get_data_rating(product_id,sp_id,page=1,limit=5):
    seller_id = get_sellerid(product_id,sp_id)
    url = 'https://tiki.vn/api/v2/reviews'
    try:

        headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
                'Accept': 'application/json',
                # Thêm các tiêu đề khác nếu cần, ví dụ như Authorization
            }
        params = {
                'limit': limit,
                'page': page,
                'include': 'comments,contribute_info,attribute_vote_summary',
                'product_id': product_id,
                'sp_id': sp_id,
                'seller_id': seller_id
            }
        # Gửi yêu cầu GET tới API
        response = requests.get(url,params=params,headers=headers)
        # Kiểm tra mã trạng thái của phản hồi
        response.raise_for_status()
        
        # Xử lý dữ liệu phản hồi
        data = response.json() 
        return data["rating_average"], data["reviews_count"]
      # In dữ liệu ra hoặc xử lý tùy ý
    except requests.exceptions.RequestException as e:
        # Xử lý lỗi
        print("Có lỗi xảy ra:", e)
```

```python
def get_all_comments(product_id,sp_id):
    seller_id = get_sellerid(product_id,sp_id)
    list_comment = []
    page = 1
    limit= 20
    while(True):
        data = get_comment(product_id,sp_id,seller_id,page,limit)
        if data == []:
            break
        for dt in data:
            comment = {}
            comment["title"] = dt["title"]
            comment["content"] = dt["content"]
            comment["rating"] = dt["rating"]
            comment["customer_id"] = dt["customer_id"]
            comment["created_at"] = dt["created_at"]
            list_comment.append(comment)    
        page = page+1
        
    return list_comment
```

```python
import csv
import os
import math

filename = 'output.csv'

# Đọc toàn bộ dữ liệu từ tệp gốc
with open(filename, mode='r', encoding='utf-8') as file:
    reader = list(csv.reader(file))
    header = reader[0]  # Lưu hàng tiêu đề
    rows = reader[1:]   # Lấy tất cả các hàng dữ liệu sau tiêu đề

# Tính toán số lượng hàng trong mỗi tệp con
num_files = 18
rows_per_file = math.ceil(len(rows) / num_files)  # Số hàng mỗi tệp

# Tạo từng tệp con và ghi dữ liệu vào
for i in range(num_files):
    # Xác định phạm vi hàng cho mỗi tệp con
    start = i * rows_per_file
    end = start + rows_per_file
    rows_subset = rows[start:end]
    
    # Tạo tệp con mới với tên chứa chỉ số
    output_filename = "divided_data//"+f'{os.path.splitext(filename)[0]}_part_{i+1}.csv'
    with open(output_filename, mode='w', encoding='utf-8', newline='') as output_file:
        writer = csv.writer(output_file)
        writer.writerow(header)  # Ghi tiêu đề vào mỗi tệp con
        writer.writerows(rows_subset)  # Ghi dữ liệu vào tệp con

print(f"Tệp '{filename}' đã được chia thành {num_files} tệp con.")
```

```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
def unique_values(data):
    total = data.count()
    tt = pd.DataFrame(total)
    tt.columns = ['Total']
    uniques = []
    for col in data.columns:
        unique = data[col].nunique()
        uniques.append(unique)
    tt['Uniques'] = uniques
    return tt
```
